```python
def monteCarloConvexity1(images,number=50000):
    counter = 0
    counterPositive12 = 0
    A12=[]

    pointX=[]
    pointY=[]
    for i in range(images.shape[0]):
        for j in range(images.shape[1]):
            if images[i][j]==1:
                pointX.append(i)
                pointY.append(j)
    if len(pointX)==0:
        return 0
    
    for i in range(number):
        p1 = np.random.randint(0,len(pointX))
        p2 = np.random.randint(0,len(pointX))
        x1 = pointX[p1]
        y1 = pointY[p1]
        x2 = pointX[p2]
        y2 = pointY[p2] 
        x12 = int((x1+x2)/2)
        y12 = int((y1+y2)/2)

        counter=counter+1
        if images[x12][y12]==1:
            counterPositive12 = counterPositive12+1
            

    return (counterPositive12/counter)
```

```python
def shortRangeConvexity(image,number=50000,sideparts = 4):
    Ab = []
    j = 0
    
    sideLength = int(image.shape[0]/sideparts)
    convexitySum=0
    
    for ii in range(sideparts):
        for jj in range(sideparts):
            img_new = image[sideLength*ii:sideLength*(ii+1),sideLength*jj:sideLength*(jj+1)]
            if np.sum(img_new)==0:
                continue
            Ab.append(monteCarloConvexity1(img_new,int(2*number/sideparts/sideparts)))
    convexityShortRange = np.mean(Ab)
    
    
    return convexityShortRange
    
```

```python
AA = []
j = 0
for i in list_files:
    j = j+1
    if j%40==0:
        img_ = corr.dat_to_numpy(i)
        img_ = vel.gaussian_filter(img_,4)
        img_ = (img_>0.5)*1
        convexity = shortRangeConvexity(img_, number = 50000, sideparts = 4)
        AA.append(convexity)
        print(convexity)
        plt.imshow(img_)
        plt.show()
```
