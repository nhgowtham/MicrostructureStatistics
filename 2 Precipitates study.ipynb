```python
path = '/home/sushil/Desktop/adithya/ISOTROPIC/C03/'
list_of_files = os.listdir(path)
list_of_files.sort()
list_files = []
for i in list_of_files:
    if i[-5:]=='0.dat':
        list_files.append(path+i)
        
img_1 = corr.dat_to_numpy(list_files[167])
img_1 = vel.gaussian_filter(img_1,5)
img_1 = (img_1>0.5)*1
start= time.time()
labels = hosh.hoshenKoplemanLabels(img_1)
print(time.time()-start, ' seconds')
plt.rcParams["figure.figsize"] = (5,5)
plt.imshow(labels)
plt.colorbar()
print('total number of ppts :', np.max(labels))
```

```python
cog = hosh.precipitateCentres(labels,93)
cog_2 = hosh.precipitateCentres(labels,102)
```

```python
# define cog here
area = []
area2 = []

cog1 = []
cog2 =[]
cog1_2=[]
cog2_2=[]

theta_array = []
l_array =[]
b_array = []
eig_array =[]

theta_array_2 = []
l_array_2 =[]
b_array_2 = []
eig_array_2 =[]


numberOfPpts = []

for i in range(50):
    start = time.time()
    img_track1 = corr.dat_to_numpy(list_files[167-i])
    img_track1 = vel.gaussian_filter(img_track1,4)
    img_track1 = (img_track1>0.5)*1
    labels_track1 = hosh.hoshenKoplemanLabels(img_track1)
    numberOfPpts.append(np.max(labels_track1))
    
    # For upper
    label_for_our_ppt = labels_track1[cog[0]][cog[1]]
    if(label_for_our_ppt==0):
        print('the microstructure dissapeared, or moved very quickly')
        break
    image_tracked1_binary = (labels_track1==label_for_our_ppt)*1
    cog = hosh.precipitateCentres(image_tracked1_binary,1)
    theta, l, b, eig = hosh.findAngleMajorMinorEigenvector(image_tracked1_binary,1)
    aa = np.sum(image_tracked1_binary)
    area.append(aa)
    cog1.append(cog[0])
    cog2.append(cog[1])
    theta_array.append(theta)
    l_array.append(l)
    b_array.append(b)
    eig_array.append(eig)
    
    # For lower
    label_for_our_ppt_2 = labels_track1[cog_2[0]][cog_2[1]]
    if(label_for_our_ppt_2==0):
        print('the microstructure 2 dissapeared, or moved very quickly')
        break
    image_tracked1_binary_2 = (labels_track1==label_for_our_ppt_2)*1
    cog_2 = hosh.precipitateCentres(image_tracked1_binary_2,1)
    theta2, l2, b2, eig2 = hosh.findAngleMajorMinorEigenvector(image_tracked1_binary_2,1)
    aa2 = np.sum(image_tracked1_binary_2)
    area2.append(aa2)
    cog1_2.append(cog_2[0])
    cog2_2.append(cog_2[1])
    theta_array_2.append(theta2)
    l_array_2.append(l2)
    b_array_2.append(b2)
    eig_array_2.append(eig2)
    
    
    
    
    plt.imshow(image_tracked1_binary+image_tracked1_binary_2)
    plt.title(str(167-i))
    plt.show()
    
    print('----------------------------------------------------------------------------')
```

```python
df = pd.DataFrame()
df['area'] = area
df['cog1'] = cog1
df['cog2'] = cog2
df['theta_array'] =theta_array 
df['l_array'] = l_array
df['b_array'] = b_array
df['eig_array'] = eig_array
df['numberOfPpts'] = numberOfPpts
df.to_pickle('2_Precipitates_study.pkl')
np.save('2_Precipitates_study_eig.npy',eig_array)
```
