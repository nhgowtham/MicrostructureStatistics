```python
import numpy as np
from PIL import Image
from Scripts import SpatialCorrelations as corr
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pymks
import pandas as pd
import os
from Scripts import velocityCalculations as vel
from Scripts import hoshenKopleman as hosh
import math
import time
from numpy import linalg as LA
from Scripts import plots
```

```python
radiusVector = np.zeros((2048,2048))
for i in range(radiusVector.shape[0]):
    for j in range(radiusVector.shape[1]):
        radiusVector[i][j] = np.sqrt((i-radiusVector.shape[0]/2)**2 +(j-radiusVector.shape[1]/2)**2)
```

```python
plt.imshow(radiusVector)
```

```python
# C03

path = '/home/sushil/Desktop/adithya/ISOTROPIC/C03/'
list_of_files = os.listdir(path)
list_of_files.sort()
list_files = []
for i in list_of_files:
    if i[-5:]=='0.dat':
        list_files.append(path+i)

microStrNoList = []
autoCorr=[]


j = 0
for i in list_files:
    if j==3 or j%30==0 or j==10 or j==20:
        microStrNoList.append(j)
        start= time.time()
        print('--------------------------------Start-------------------------------------------')
        print('MS number : ',j)
        
        print('--------Auto Correlation below--------')
        img_ = corr.dat_to_numpy(i)
        img_ = vel.gaussian_filter(img_,4)
        img_ = (img_>0.5)*1
        white_stats, black_stats = corr.auto_corr_from_code(img_)
        
        
        print('-------Radial Auto Correlation HM------')
        r_val =corr.radialDistribution_hm(white_stats,radiusVector)
        autoCorr.append(r_val)
        print('time = ',time.time()-start)
        print('-----------------------------------End------------------------------------------')

    j=j+1

np.save('/home/sushil/Desktop/adithya/code/MicrostructureStatistics/IsotropicPlotData/autoCorr_hm.npy',np.array(autoCorr))
   
```
